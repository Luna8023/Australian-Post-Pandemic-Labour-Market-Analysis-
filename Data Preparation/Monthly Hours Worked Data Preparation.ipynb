## Import libraries and datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df1 = pd.read_excel('Monthly hours worked by Employment Type by Sex and by State.xls', sheet_name='Data1', index_col=0, parse_dates=True)
df2 = pd.read_excel('Monthly hours worked in all jobs by employment type by Sex and Territory.xls', sheet_name='Data1', index_col=0, parse_dates=True)
```

## Preparing Dataset

```python
df1.head(10)
```

```python
df2.head()
```

The original dataset in df1 contain both seasonally adjusted data and the trend, we will use Seasonally Adjusted data here and drop the trend before concatenating with df2.

```python
# Given all the Seasonally Adjusted data in df1 contain ';.1' in the column name, we will use this feature to filter these columns
df1 = df1.filter(like=';.1', axis = 1)
# Alternatively, we can use df.loc[:, df.columns.str.contains('1')]
```

```python
df= pd.concat([df1, df2], axis=1)
```

```python
df.head()
```

```python
# After filtering the selected columns, the index 'Unit', 'Series Type' and etc are not needed
df.drop(index=['Unit', 'Series Type', 'Data Type', 'Frequency', 'Collection Month', 'Series Start', 
               'Series End', 'No. Obs', 'Series ID'], axis=0, inplace=True)
```

```python
df.head()
```

```python
# Fixing messy column names
# Remove unwanted ';' and '>' in the column name
df.columns = df.columns.str.strip().str.replace(';  ', '').str.replace('> ', '').str.replace(' ;.1', '(000)').str.replace(' ;', '(000)').str.replace('Persons', '')
# Remove extra space in the middle and at the end
df.columns = df.columns.str.rstrip().str.replace('  ', '')
```

```python
df.columns
```

```python
# Generate new columns for Year and Month
df.index = pd.to_datetime(df.index)
year = df.index.year
month = df.index.month 
df.insert(0, 'Year', year)
df.insert(1, 'Month', month)
```

```python
# Check if column names are more organised
df.head()
```

```python
df.dtypes
```

```python
# Convert all the columns from object to float or int with one touch
df = df.infer_objects()
```

```python
df['Monthly worked hour growth (000)'] = df['Monthly hours worked in all jobs (000)'].diff()
df['Monthly worked hour growth Males (000)'] = df['Monthly hours worked in all jobs Males(000)'].diff()
df['Monthly worked hour growth Females (000)'] = df['Monthly hours worked in all jobs Females(000)'].diff()
df['Monthly full-time worked hour growth (000)'] = df['Monthly hours worked in all jobs (employed full-time) (000)'].diff()
df['Monthly part-time worked hour growth (000)'] = df['Monthly hours worked in all jobs (employed part-time) (000)'].diff()
```

To calculate the worked hours per thousand people in each state, the state population data was obtained from ABS website. The data showed the population on 30 Sep 2019, the unit is (000). 
https://www.abs.gov.au/ausstats/abs@.nsf/Latestproducts/3101.0Main%20Features3Sep%202019?opendocument&tabname=Summary&prodno=3101.0&issue=Sep%202019&num=&view= <br>
New South Wales	8 118.0 <br>
Victoria	6 629.9 <br>
Queensland	5 115.5 <br>
South Australia	1 756.5<br>
Western Australia	2 630.6 <br>
Tasmania	535.5 <br>
Northern Territory	245.6 <br>
Australian Capital Territory	428.1 <br>

```python
# Calculate monthly worked hours per thousand people. 
df['Monthly hours worked per thousand people New South Wales(000)'] = df['Monthly hours worked in all jobs New South Wales(000)']/8118.0 
df['Monthly hours worked per thousand people Victoria(000)'] = df['Monthly hours worked in all jobs Victoria(000)']/6629.9 
df['Monthly hours worked per thousand people Queensland(000)'] = df['Monthly hours worked in all jobs Queensland(000)']/5115.5 
df['Monthly hours worked per thousand people South Australia(000)'] = df['Monthly hours worked in all jobs South Australia(000)']/1756.5 
df['Monthly hours worked per thousand people Western Australia(000)'] = df['Monthly hours worked in all jobs Western Australia(000)']/2630.6 
df['Monthly hours worked per thousand people Tasmania(000)'] = df['Monthly hours worked in all jobs Tasmania(000)']/535.5
df['Monthly hours worked per thousand people Northern Territory(000)'] = df['Monthly hours worked in all jobs Northern Territory(000)']/245.6
df['Monthly hours worked per thousand people Australian Capital Territory(000)'] = df['Monthly hours worked in all jobs Australian Capital Territory(000)']/428.1
```
