## Import libraries and datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df1 = pd.read_excel('Monthly hours worked by Employment Type by Sex and by State.xls', sheet_name='Data1', index_col=0, parse_dates=True)
df2 = pd.read_excel('Monthly hours worked in all jobs by employment type by Sex and Territory.xls', sheet_name='Data1', index_col=0, parse_dates=True)
```

## Preparing Dataset

```python
df1.head(10)
```

```python
df2.head()
```

The original dataset in df1 contain both seasonally adjusted data and the trend, we will use Seasonally Adjusted data here and drop the trend before concatenating with df2.

```python
# Given all the Seasonally Adjusted data in df1 contain ';.1' in the column name, we will use this feature to filter these columns
df1 = df1.filter(like=';.1', axis = 1)
# Alternatively, we can use df.loc[:, df.columns.str.contains('1')]
```

```python
df= pd.concat([df1, df2], axis=1)
```

```python
df.head()
```

```python
# After filtering the selected columns, the index 'Unit', 'Series Type' and etc are not needed
df.drop(index=['Unit', 'Series Type', 'Data Type', 'Frequency', 'Collection Month', 'Series Start', 
               'Series End', 'No. Obs', 'Series ID'], axis=0, inplace=True)
```

```python
df.head()
```

```python
# Fixing messy column names
# Remove unwanted ';' and '>' in the column name
df.columns = df.columns.str.strip().str.replace(';  ', '').str.replace('> ', '').str.replace(' ;.1', '').str.replace('Persons', '')
# Remove extra space in the middle and at the end
df.columns = df.columns.str.rstrip().str.replace('  ', '')
```

```python
df.columns
```

```python
# Generate new columns for Year and Month
df.index = pd.to_datetime(df.index)
year = df.index.year
month = df.index.month 
df.insert(0, 'Year', year)
df.insert(1, 'Month', month)
```

```python
# Check if column names are more organised
df.head()
```

```python
df.dtypes
```

```python
# Convert all the columns from object to float or int with one touch
df = df.infer_objects()
```

```python
df['Monthly worked hour growth'] = df['Monthly hours worked in all jobs'].diff()
df['Monthly worked hour growth Males'] = df['Monthly hours worked in all jobs Males'].diff()
df['Monthly worked hour growth Females'] = df['Monthly hours worked in all jobs Females'].diff()
df['Monthly full-time worked hour growth'] = df['Monthly hours worked in all jobs (employed full-time)'].diff()
df['Monthly part-time worked hour growth'] = df['Monthly hours worked in all jobs (employed part-time)'].diff()
```
